## between method

```python
# get house between 5 and 7 bedroom
houses['bedrooms'].between(5, 7).value_counts()
```

## isin() method

```python
bedrooms = [1,2,3]
houses["bedrooms"].isin(bedrooms)
```

```python
houses[houses['bedrooms'].isin(bedrooms)]
```

## Combine multiple condition

```python
# find condition with the houses at the waterfront, 3 bedroom, and affordable
waterfront = houses["waterfront"] == 1
waterfront
```

```python
threebedrooms = houses["bedrooms"] == 3
threebedrooms
```

```python
affordable = houses['price'] < 600000
affordable
```

```python
houses[waterfront & threebedrooms & affordable]
```

```python
# get a dream home
high_quality = houses["grade"] >= 11
good_view = houses["view"] == 4
smaller = houses["sqft_living"] <= 3000
houses[high_quality & good_view & smaller]
```

```python
# find newer condition home using OR

houses[houses["yr_built"] >= 2014]
```

```python
houses[houses["yr_renovated"] >= 2014]
```

```python
newer_condition = houses[(houses["yr_built"] >= 2014) | (houses["yr_renovated"] >= 2014)]
```

```python
newer_condition.sort_values("yr_renovated")
```

## The tilt method is a good way to get a flip of the opposite of what we select. 

```python
newly_built = houses["yr_built"] >= 2014
newly_renovate = houses["yr_renovated"] >= 2014
modern_homes = newly_built | newly_renovate
houses[modern_homes]
```

```python
# older home
houses[~modern_homes]
```

```python
# people is not survived
tinanic[tinanic.survived == 0]

```

```python
# people is survide
tinanic[~(tinanic.survived == 0)]
```

```python
women = tinanic["sex"] == "female"
tinanic[women]["survived"].value_counts().plot(kind="pie")
```

```python
# male
tinanic[~women]["survived"].value_counts().plot(kind="pie")
```

# Filter exercise

```python
import pandas as pd 
books = pd.read_csv("data/bestsellers.csv")
books.head()
```

### find the books written by Pete Souza

```python
books[books["Author"] == "Pete Souza"]
```

### find the books that are under 10 dollars

```python
books[books["Price"] < 10]
```

### find the books that have a price between 50 and 60 dollars

```python
books[books["Price"].between(50, 60)]
```

### find all the books written by Kristin Hannah, Andy Weir, or Delia Owens

```python
authors = ["Kristin Hannah", "Andy Weir", "Delia Owens"]
books[books["Author"].isin(authors)]
```

### Find the Non Fiction books that are rated 4.9

```python
highRatedNonfiction = (books["Genre"] == "Non Fiction") & (books["User Rating"] == 4.9)
books[highRatedNonfiction]
```

### Find the fiction book with the lowest User Rating

```python
books[books["Genre"] == "Fiction"].sort_values("User Rating")
```

### Find 2012's top 5 Fiction books with the most Reviews

```python
df = books[(books["Year"] == 2012) & (books["Genre"] == "Fiction")]
df.sort_values("Reviews", ascending=False).head()
```

### Create a bar plot showing the 5 authors who have the most books with the rating under 4.5

```python
low_rated = books[books["User Rating"] < 4.5]
low_rated["Author"].value_counts().head().plot(kind="bar")
```
